```python
import pandas as pd
import csv
from docx import Document

def docx_to_csv(docx_filename, csv_filename):
    # Open the DOCX file
    doc = Document(docx_filename)
    
    # Prepare to write to the CSV file
    with open(csv_filename, mode='w', newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)
        
        # Iterate through the paragraphs in the DOCX file
        for para in doc.paragraphs:
            writer.writerow([para.text])
        
        # Iterate through the tables in the DOCX file
        for table in doc.tables:
            for row in table.rows:
                row_data = [cell.text for cell in row.cells]
                writer.writerow(row_data)

# Convert DOCX to CSV
docx_filename = 'climate.docx'
csv_filename = 'climate.csv'
docx_to_csv(docx_filename, csv_filename)

# Now read the CSV file using pandas
df = pd.read_csv(csv_filename)
print("Columns in the dataset:", df.columns)
print(df.head())
```

```python
# app.py

from flask import Flask, request, jsonify
from joblib import load
import numpy as np

app = Flask(__name__)

# Load the pre-trained model
model = load('flood_model.pkl')

@app.route('/')
def home():
    return "Flood Prediction Model API"

@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Get data from the request
        data = request.get_json(force=True)

        # Extract features from data
        rainfall = data.get('rainfall')
        river_level = data.get('river_level')

        # Validate input data
        if rainfall is None or river_level is None:
            return jsonify({'error': 'Invalid input data'}), 400

        # Prepare features for prediction
        features = np.array([rainfall, river_level]).reshape(1, -1)

        # Make prediction
        prediction = model.predict(features)

        # Return the result as JSON
        result = {'prediction': int(prediction[0])}
        return jsonify(result)

    except Exception as e:
        return jsonify({'error': str(e)}), 500

if __name__ == '__main__':
    app.run(debug=True)
```
